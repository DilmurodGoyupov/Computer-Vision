```python
# zoom_image function. (arr-image, n-fold magnification)
def zoom_image(arr,n):
    
    for i in range(n-1):
        #print(a)
        arr=arr.swapaxes(1,0)
        print()
        arr=arr.repeat(2,axis=1)
        #print(a)
        arr=arr.repeat(2,axis=0)
        #print(a)
        arr=arr.swapaxes(0,1)
       
    return arr
```

```python
n=int(input('n='))

rasm_zoom=zoom_image(rasm,n)

cv2.imshow("rasm",rasm_zoom)     #show the picture
cv2.waitKey(0)
cv2.destroyAllWindows()
```
