# Required libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("Photo/sel.png")     #load images from local file
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb 
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(img)     #show the picture
```

```python
back=cv2.imread("Photo/back2.jpg")     #Download the green background image from the local file
back=cv2.cvtColor(back,cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(back)     #show the picture
```

```python
range1=np.array([3,116,0])     #first range
range2=np.array([172,255,170])     #second range

#Create mask
maska=cv2.inRange(img,range1,range2)
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(maska,cmap='gray')     #show mask
```

# Range1 = 3,116,0



# Range2 = 172,255,170


# Range1 and Range2 help us to hide the green values in this range


```python
plt.figure(figsize=(15,10))

#show original image
plt.subplot(131)
plt.title("Original image")
plt.imshow(img)

#show background image
plt.subplot(132)
plt.title("Background image")
plt.imshow(back)

#show mask image
plt.subplot(133)
plt.title("Maska")
plt.imshow(maska,cmap='gray')
```

```python
#Size original image
img.shape
```

```python
#Size mask image
maska.shape
```

```python
#Size background image
back.shape
```

```python
#Resize the background image to the size of the original image
back=cv2.resize(back, (737,818))
back.shape
```

```python
#in mask values equal to 255, we set the values of the original image to 0
img[maska==255]=[0,0,0]
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(img)
```

```python
# at the values of the mask equal to 255, we set the values of the background image to 0
back[maska==0]=[0,0,0]
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(back)
```

```python
plt.figure(figsize=(15,10))

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(back)
```

## Result

```python
final_img=img+back
plt.figure(figsize=(7,7))  #the size of the frame that displays the image
plt.imshow(final_img)
```
