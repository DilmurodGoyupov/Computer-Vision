# <center> Required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# <center> Picture normalization

```python
img=cv2.imread("Photo/tash.jpg")     #load images from local file

img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb 
plt.figure(figsize=(10,10))  #the size of the frame that displays the image
plt.imshow(img)     #show the picture
```

```python
print(img)     #show picture matrix
```

```python
img=img.astype("float")/255.0     #To normalize the image, we need to divide each matrix value by 255
plt.figure(figsize=(10,10))
plt.imshow(img)     # by dividing by 255 we do not lose quality
```

```python
print(img)     # show picture matrix [0...1]
```

# <center> Quality reduction

```python
img_reduce=np.around(img,1)     #once our images are between 0 and 1, we shrink them to one after the dot using np.around
plt.figure(figsize=(10,10))
plt.imshow(img_reduce) # view low quality image
```

```python
print(img_reduce)     #show quality reduction matrix
```
