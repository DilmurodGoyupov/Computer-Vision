```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
input_img = cv2.imread("Photo/d.png")    #load images from local file
plt.imshow(input_img)      #show the picture
```

# Dilate

` Increases the object area `

* A kernel(a matrix of odd size(3,5,7) is convolved with the image

* A pixel element in the original image is ‚Äò1‚Äô if at least one pixel under the kernel is ‚Äò1‚Äô.

* It increases the white region in the image or the size of the foreground object increases¬†


<a href=" https://www.youtube.com/watch?v=3IJ8RFtlDLY">The working principle of dilate</a>

```python
kernel = np.ones((7,7), np.uint8)
dilate_img =  cv2.dilate(input_img,kernel,iterations=1)
plt.imshow(dilate_img) # show dilate picture
```

# Erode

` Used to diminish the features of an image. `

* A kernel(a matrix of odd size(3,5,7) is convolved with the image.

* A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel are 1, otherwise, it is eroded (made to zero).

* Thus all the pixels near the boundary will be discarded depending upon the size of the kernel.

* So the thickness or size of the foreground object decreases or simply the white region decreases in the image.


<a href=" https://www.youtube.com/watch?v=b5lgnNEzGeU">The working principle of erode</a>


```python
kernel = np.ones((7,7), np.uint8)
erode_img = cv2.erode(input_img, kernel, iterations=1)
plt.imshow(erode_img) # show erode picture
```

```python
plt.figure(figsize=(15,15))

plt.subplot(131)
plt.title("Input Image")
plt.imshow(input_img)

plt.subplot(132)
plt.title("Dilate Image")
plt.imshow(dilate_img)

plt.subplot(133)
plt.title("Erode Image")
plt.imshow(erode_img)
```
