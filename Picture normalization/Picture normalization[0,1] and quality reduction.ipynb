# Required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Picture normalization

```python
rasm=cv2.imread("Photo/tash.jpg")     #load images from local file

rasm=cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb 

plt.imshow(rasm)     #show the picture
```

```python
print(rasm)     #show picture matrix
```

```python
rasm=rasm.astype("float")/255.0     #To normalize the image, we need to divide each matrix value by 255

plt.imshow(rasm)     # by dividing by 255 we do not lose quality
```

```python
print(rasm)     # show picture matrix [0...1]
```

# Quality reduction

```python
rasm_reduce=np.around(rasm,1)     #once our images are between 0 and 1, we shrink them to one after the dot using np.around
plt.imshow(rasm_reduce) # view low quality image
```

```python
print(rasm_reduce)     #show quality reduction matrix
```
