```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Sample Image

```python
sample_img = np.random.randint(0,10, size=(8,8), dtype='uint8')
print(sample_img)
plt.imshow(sample_img)
```

# Zoom out

* Inter Nearest

` [0][0] gets the index, [0][1] drops the index, [0][2] gets the index, [0][3] drops the index ...... `

` [0][0] gets the index, [1][0] drops the index, [2][0] gets the index, [3][0] drops the index ...... `

```python
zoom_out_nearest = cv2.resize(sample_img, (4,4), interpolation=cv2.INTER_NEAREST)
print(sample_img,"\n")

print(zoom_out_nearest)
plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Nearest Image")
plt.imshow(zoom_out_nearest)
```


* Inter Linear

` [0][0]+[0][1]+[1][0]+[1][1] (indices change accordingly [0][2]+[0][3]+[1 ][2]+[1][3]) adds the numbers and takes the average and rounds up to the nearest `

```python
zoom_out_linear = cv2.resize(sample_img, (4,4), interpolation=cv2.INTER_LINEAR)
print(sample_img,'\n')

print(zoom_out_linear)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Linear Image")
plt.imshow(zoom_out_linear)
```

* Inter Area

` INTER_AREA [0][0]+[0][1]+[1][0]+[1][1] (indices change accordingly [0][2]+[0][3]+[1 ][2]+[1][3])
adds the numbers and takes the average and rounds up to the nearest. `

` INTER_AREA and INTER_LINEAR work the same for reduction `

```python
zoom_out_area = cv2.resize(sample_img, (4,4), interpolation=cv2.INTER_AREA)
print(sample_img,'\n')

print(zoom_out_area)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Area Image")
plt.imshow(zoom_out_area)
```

* Inter Cubic
`bicubic interpolation over 4√ó4 pixel neighborhood`

```python
zoom_out_cubic = cv2.resize(sample_img, (4,4), interpolation=cv2.INTER_CUBIC)
print(sample_img,'\n')

print(zoom_out_cubic)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Cubic Image")
plt.imshow(zoom_out_cubic)
```

# Zoom in

* Inter Nearset 

` Each pixel is surrounded by the same pixel `

```python
zoom_in_nearest = cv2.resize(sample_img, (16,16), interpolation=cv2.INTER_NEAREST)
print(sample_img,"\n")

print(zoom_in_nearest)
plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Nearest Image")
plt.imshow(zoom_in_nearest)
```

* Inter Linear - bilinear interpolation 


```python
zoom_in_linear = cv2.resize(sample_img, (16,16), interpolation=cv2.INTER_LINEAR)
print(sample_img,'\n')

print(zoom_in_linear)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Linear Image")
plt.imshow(zoom_in_linear)
```

* Inter Area - esampling using pixel area ratio

```python
zoom_in_area = cv2.resize(sample_img, (16,16), interpolation=cv2.INTER_AREA)
print(sample_img,'\n')

print(zoom_in_area)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Area Image")
plt.imshow(zoom_in_area)
```

* Inter Cubic - bicubic interpolation in the vicinity of 4x4 pixels

```python
zoom_in_cubic = cv2.resize(sample_img, (16,16), interpolation=cv2.INTER_CUBIC)
print(sample_img,'\n')

print(zoom_in_cubic)

plt.figure(figsize=(5,5))

plt.subplot(121)
plt.title("Sample Image")
plt.imshow(sample_img)

plt.subplot(122)
plt.title("Inter Cubic Image")
plt.imshow(zoom_in_cubic)
```
