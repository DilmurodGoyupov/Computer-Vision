## <center> We saw in (4) how to work with a green background image. We now want to consider a similar situation. But our green background is slightly different, that is, there is a shadow. To solve this problem, we can read the image in HSV format. And we compare the last result with the RGB image. 

## RGB

R - Red channel

G - Green channel

B - Blue chennel



## HSV

H - HUE (COLOR (for example, red, green or blue))

S - Saturation

V - Value (light)



```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# <center> Load RGB Image

```python
img_rgb = cv2.imread("Photo/car_green.jpg")     #load images from local file

img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb 

plt.figure(figsize=(10,10))

plt.imshow(img_rgb)     #show the picture
```

```python
range1=np.array([1,48,3])     #first range
range2=np.array([153,246,155])     #second range

#Create mask
maskargb=cv2.inRange(img_rgb,range1,range2)

plt.figure(figsize=(10,10))

plt.imshow(maskargb,cmap='gray')     #show mask
```

## Range1


## Range2


## Range1 and Range2 help us to hide the green values in this range


# <center> Load HSV Image

```python
img_hsv = cv2.imread("Photo/car_green.jpg")     #load images from local file

img_hsv = cv2.cvtColor(img_hsv, cv2.COLOR_BGR2HSV)     #convert image from hsv to rgb

plt.figure(figsize=(10,10))

plt.imshow(img_hsv)     #show the picture
```

```python
range11=np.array([50,0,0])     #first range
range22=np.array([100,255,255])     #second range

#Create mask
maskahsv=cv2.inRange(img_hsv,range11,range22)
plt.figure(figsize=(10,10))
plt.imshow(maskahsv,cmap='gray')     #show mask
```

## Range1



## Range2


```python
plt.figure(figsize=(15,10))

#show rgb image mask
plt.subplot(221)
plt.title("RGB maska")
plt.imshow(maskargb,cmap='gray')

#show hsv image mask
plt.subplot(222)
plt.title("HSV maska")
plt.imshow(maskahsv,cmap='gray')

#show resalt rgb
img_rgb[maskargb==255]=0
plt.subplot(223)
plt.title("RGB Result")
plt.imshow(img_rgb)

#show resalt hsv
img_hsv[maskahsv==255]=0
img_hsv = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
plt.subplot(224)
plt.title("HSV Result")
plt.imshow(img_hsv)
```

## <center> If you look between the wheels, you can see the difference, if we get green in the RGB image through the range, we can't get all the saturation, if we get the Hsv image, we can enter the color through H, the saturation through S, and the value through V. It is better to use hsv in images with saturation. It's better to use RGB for non-saturation images

```python
plt.figure(figsize=(15,15))

plt.subplot(121)
plt.imshow(img_rgb)

plt.subplot(122)
plt.imshow(img_hsv)
```
