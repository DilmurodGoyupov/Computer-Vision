```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# The junction of two edges is called a corner

```python
img = cv2.imread("Photo/kvad.png")      #load images from local file
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)      #convert image from bgr to rgb 
plt.figure(figsize=(8,8))
plt.imshow(img)     #show the picture
```

```python
img_gray=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)     # we create an image in grayscale.
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap='gray')
```

```python
img_gray=np.float32(img_gray)     # transfer the gray image to np.float 32
corner = cv2.cornerHarris(img_gray,5,7,0.04)  
plt.figure(figsize=(8,8))
plt.imshow(corner,cmap='gray')



# corner = cv2.cornerHarris(img_gray,5,7,0.04) = line 2
# 5 block size is the size of the corner (actually the whites) that we found
# 7 kernel size
# 0.04 harris free paramet (min=0.04, max=0.07  recommended)
```

```python
corner.shape
```

```python
for i in range(len(corner[0])):
    for j in range(len(corner[1])):
        print(corner[i][j], end=' ')
    print()
```

```python
print(corner.max())
```

```python
print(corner.min())
```

# Defining corners

```python
threshold = 0.5*corner.max()

for i in range(0, corner.shape[0]):
    for j in range(0, corner.shape[1]):
        if corner[i,j]>threshold:
            cv2.circle(img, (j,i), 1, (255,0,0),5)
plt.figure(figsize=(8,8))
plt.imshow(img,cmap='gray')
```

# <center> <h2>  Harris Corner Detection Algorithm

```python
img = cv2.imread("Photo/kvad.png")
plt.figure(figsize=(8,8))
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray, cmap="gray")
```

```python
height = img.shape[0]   #.shape[0] outputs height 
height
```

```python
width = img.shape[1]    #.shape[1] outputs width .shape[2] outputs color channels of image
width
```

```python
matrix_R = np.zeros((height,width))
matrix_R
```

```python
Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3) 
Ix
# cv2.Sobel allows us to find derivatives, 1, 0 is along the x-axis
```

```python
plt.figure(figsize=(8,8))
plt.imshow(Ix, cmap="gray")
```

```python
Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  
Iy
# cv2.Sobel allows us to find derivatives, 0, 1 is along the y-axis
```

```python
plt.figure(figsize=(8,8))
plt.imshow(Iy, cmap="gray")
```

```python
Ix2=np.square(Ix)
Ix2
```

```python
plt.figure(figsize=(8,8))
plt.imshow(Ix2, cmap="gray")
```

```python
Iy2=np.square(Iy)
Iy2
```

```python
plt.figure(figsize=(8,8))
plt.imshow(Iy2, cmap="gray")
```

```python
Ixy=Ix*Iy
Ixy
```

```python
offset = int( 5 / 2 )  #int( window_size / 2 )
offset
```

```python
print ("Finding Corners...")
k=0.004
for y in range(offset, height-offset):
    for x in range(offset, width-offset):
        
        #Step 3
        
        Sx2 = np.sum(Ix2[y-offset:y+1+offset, x-offset:x+1+offset])
        Sy2 = np.sum(Iy2[y-offset:y+1+offset, x-offset:x+1+offset])
        Sxy = np.sum(Ixy[y-offset:y+1+offset, x-offset:x+1+offset])

        #   Step 4
        
        H = np.array([[Sx2,Sxy],[Sxy,Sy2]])

        #   Step 5
       
        det=np.linalg.det(H)
        tr=np.matrix.trace(H)
        R=det-k*(tr**2)   #k=0.004
        matrix_R[y-offset, x-offset]=R
```

```python
threshold = 0.30
cv2.normalize(matrix_R, matrix_R, 0, 1, cv2.NORM_MINMAX)

for y in range(offset, height-offset):
    for x in range(offset, width-offset):
        value=matrix_R[y, x]
        
        if value>threshold:
            # cornerList.append([x, y, value])
            cv2.circle(img,(x,y),3,(0,0,255))
```

```python
plt.figure(figsize=(8,8))

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title("Manually implemented Harris detector")

plt.show()
```
