# Required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Photo/tash.jpg')     #load images from local file
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #convert image from bgr to rgb 
plt.figure(figsize=(8,8))   #the size of the frame that displays the image
plt.imshow(img) #show the picture
```

```python
img.shape     #view image size
```

```python
# zoom_image function. (arr-image, n-fold magnification)
def zoom_image(arr,n):
    
    for i in range(n-1):
        #print(a)
        arr=arr.swapaxes(1,0)
        print()
        arr=arr.repeat(2,axis=1)
        #print(a)
        arr=arr.repeat(2,axis=0)
        #print(a)
        arr=arr.swapaxes(0,1)
       
    return arr
```

```python
n=int(input('n='))
```

```python
img_zoom=zoom_image(img,n)
plt.figure(figsize=(10,10))
plt.imshow(img_zoom)
```
