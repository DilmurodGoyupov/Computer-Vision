```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("Photo/tash.jpg")
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
red_chanel=img[:,:,0]
green_chanel=img[:,:,1]
blue_chanel = img[:,:,2]
```

```python
plt.figure(figsize=(15,10))

plt.subplot(131)
plt.title("Red chanel")
plt.imshow(red_chanel)

plt.subplot(132)
plt.title("Green chanel")
plt.imshow(green_chanel)

plt.subplot(133)
plt.title("Blue chanel")
plt.imshow(blue_chanel)
```
