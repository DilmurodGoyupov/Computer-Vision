```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("Photo/team.jpg")     #load images from local file
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)      #convert image from bgr to rgb 
plt.imshow(img)   
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)     # we create an image in grayscale.
plt.imshow(img_gray,cmap='gray')
```

# <center> Makes the values 0 in 0 to 188. Makes the values 188 to 255 255. Then we have an image of only 0 and 255.

```python
rat, binary_img = cv2.threshold(img_gray,188,255,cv2.THRESH_BINARY_INV)     # we transfer to a binary image
plt.imshow(binary_img,cmap='gray')     # the image consists of only 0 and 255
```

```python
for i in range(binary_img.shape[0]):
    for j in range(binary_img.shape[1]):
        print(binary_img[i][j], end=' ')
    print()

```

```python
contour, hierarchy  = cv2.findContours(binary_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE) # Find contours
print(contour)
# So now that our images are only 0's and 255's. 
# All we have to do is find the coordinates of the places where our color changed from 0 to 255 in the image.
```

 `cv.findContours() function, `

`first one is source image,`

`second is contour retrieval mode,`

`third is contour approximation method.`

And it outputs the contours and hierarchy. Contours is a Python list of all the contours in the image. Each individual contour is a Numpy array of (x,y) coordinates of boundary points of the object.

```python
print(hierarchy)

# Sometimes objects are in different locations. But in some cases, some shapes are inside other shapes.
# Just like nested figures. In this case, we call outer one as parent and inner one as child.
# This way, contours in an image has some relationship to each other. 
# And we can specify how one contour is connected to each other, like, is it child of some other contour, or is it a parent etc. 
# Representation of this relationship is called the Hierarchy.

# https://docs.opencv.org/4.x/d9/d8b/tutorial_py_contours_hierarchy.html
```

```python
rasm_contour = cv2.drawContours(img,contour,-1,(0,255,0),2)  
plt.imshow(rasm_contour)

# parametr third in cv2.drawContours (-1)
# The pixel coordinates of the contour points are listed in the obtained contours. 
# Using this argument,you can specify the index position from this list,indicating exactly which contour point you want to draw. 
#Providing a negative value will draw all the contour points.

```
