# Required libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# <center> Color images are 3D. They consist of a combination of red, blue and green colors.





```python
img=cv2.imread("Photo/tash.jpg")          #load images from local file.
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #convert image from bgr to rgb
plt.figure(figsize=(10,10))  #the size of the frame that displays the image
plt.imshow(img)  #show the picture
```

```python
red_chanel=img[:,:,0]
green_chanel=img[:,:,1]
blue_chanel = img[:,:,2]
```

```python
plt.figure(figsize=(15,10))

plt.subplot(131)          #1 - row, 3 - number of pictures (or number of columns), 1 - position of picture
plt.title("Red chanel")
plt.imshow(red_chanel)    #show red channel

plt.subplot(132)          #1 - row, 3 - number of pictures (or number of columns), 2 - position of picture
plt.title("Green chanel")
plt.imshow(green_chanel)  #green red channel

plt.subplot(133)          #1 - row, 3 - number of pictures (or number of columns), 3 - position of picture
plt.title("Blue chanel")
plt.imshow(blue_chanel)   #show blue channel
```
