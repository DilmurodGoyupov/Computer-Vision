# Face 

```python
facemodels = cv2.CascadeClassifier("Models/haarcascade_frontalface_default.xml")  # load model from local file
print(facemodels)
```

```python
face = cv2.imread("Photo/faces.jpg")     #load images from local file
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)     #convert image from bgr to rgb
plt.figure(figsize=(15,10))
plt.imshow(face)     #show the picture
plt.show()
```

```python
faces = facemodels.detectMultiScale2(face, scaleFactor=2,minNeighbors=1) 
# scaleFactor - Parameter specifying how much the image size is reduced at each image scale.
# minNeighbors - Parameter specifying how many neighbors each candidate rectangle should have to retain it.
print(faces)
#Columns 1 and 2 in the matrix are the x and y coordinates of these faces.
#Columns 3 and 4 width and height of the rectangle
```

```python
face_copy = np.copy(face)     # We copy the picture. So that we do not draw on the original picture

for x,y,w,h in faces[0]:
    cv2.rectangle(face_copy, (x,y), (x+w, y+h), (0,255,0),2)
plt.figure(figsize=(15,10))
plt.imshow(face_copy)
```

```python
eyesmodels = cv2.CascadeClassifier("Models/haarcascade_eye.xml")
print(eyesmodels)
```

```python
eyes = eyesmodels.detectMultiScale(faceeye, minSize=(10,10),maxSize=(40,40))
print(eyes)
```
