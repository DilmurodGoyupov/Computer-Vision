```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
intro = cv2.VideoCapture("Photo/kir.mp4")

while intro.isOpened():
    _, back = intro.read()
    
    back=cv2.resize(back, (640,480))
    
    cv2.imshow("Intro",back)
    
    if cv2.waitKey(25) & 0xFF==ord("q"):
        break
        
intro.release()
cv2.destroyAllWindows()
```

```python
tansor = cv2.VideoCapture("Photo/tan.mp4")

while tansor.isOpened():
    _, rasm = tansor.read()
    
    rasm=cv2.resize(rasm, (640,480))
    
    cv2.imshow("Intro",rasm)
    
    if cv2.waitKey(60) & 0xFF==ord("q"):
        break
        
tansor.release()
cv2.destroyAllWindows()
```
